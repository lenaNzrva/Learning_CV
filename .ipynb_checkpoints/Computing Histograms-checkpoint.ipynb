# Computing Histograms

Allow to visualise the distribution of the pixel intensities in an image

### For Gray

```python
import cv2 as cv
import matplotlib.pyplot as plt

img_1 = cv.imread('Photos/cats.jpg')
img_2 = cv.imread('Photos/cats 2.jpg')
gray_1 = cv.cvtColor(img_1, cv.COLOR_BGR2GRAY)
gray_2 = cv.cvtColor(img_2, cv.COLOR_BGR2GRAY)

cv.imshow('Cats 1', img_1)
cv.imshow('Cats Gray 1', gray_1)
cv.imshow('Cats 2', img_2)
cv.imshow('Cats Gray 2', gray_2)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
hist_1 = cv.calcHist([gray_1], [0], None, [256], (0, 256))
hist_2 = cv.calcHist([gray_2], [0], None, [256], (0, 256))

plt.plot(hist_1)
plt.show()

plt.plot(hist_2)
plt.show()
```

### Gray hist with mask

```python
import numpy as np
blank = np.zeros((img_2.shape[:2]), 'uint8')
circle = cv.circle(blank, (img_2.shape[1]//2, img_2.shape[0]//2), 60, 255, -1)
mask = cv.bitwise_and(gray_2, gray_2, mask = circle)
mask_hist = cv.calcHist([gray_2], [0], mask, [256], (0,256))

# cv.imshow('Blank', blank)
# cv.imshow('Circle', circle)
cv.imshow('Mask', mask)
cv.waitKey(0)
cv.destroyAllWindows()

plt.plot(mask_hist)
plt.show()
```

### Color Hist

```python
cv.imshow('Color Cats', img_1)

cv.waitKey(0)
cv.destroyAllWindows()

colors = ('b', 'g', 'r')

for i,col in enumerate(colors):
    hist = cv.calcHist([img_1], [i], None, [256], (0,256))
    plt.plot(hist, color = col)
    plt.xlim(0,255)

plt.show()
```

### Color hist with mask

```python
cv.imshow('Color Cats', img_1)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
blank = np.zeros((img_1.shape[:2]), 'uint8')
circle_mask = cv.circle(blank, (img_1.shape[1]//2, img_1.shape[0]//2), 50, 255, -1)

circle_mask_just_for_show = cv.bitwise_and(img_1, img_1, mask = circle_mask)

cv.imshow('Mask', circle_mask_just_for_show)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
colors = ('b','g','r')
for i,col in enumerate(colors):
    hist = cv.calcHist([img_1], [i], circle_mask, [256], [0,256])
    plt.plot(hist, color = col)
    
plt.show()
```
