# Color Channels

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('Photos/park.jpg')
blank = np.zeros(img.shape[:2], 'uint8')
img_rbg = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img_rbg)
plt.show()
```

```python
b,g,r = cv.split(img)
```

```python
print(img.shape)
print(b.shape)
print(g.shape)
print(r.shape)
```

```python
cv.imshow('Blue', b)
cv.imshow('Green', g)
cv.imshow('Red', r)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img_merged = cv.merge([r,g,b])
plt.imshow(img_merged)
plt.show()
```

```python
blue = cv.merge([b,blank,blank])
green = cv.merge([blank, g, blank])
red = cv.merge([blank,blank,r])

cv.imshow('Blue', blue)
cv.imshow('Red', red)
cv.imshow('Green', green)

cv.waitKey(0)
cv.destroyAllWindows()
```
