```python
import cv2 as cv
import matplotlib.pyplot as plt
```

# Read and Show photos and videos

How we can import images

```python
#Just cat image
img_1 = cv.imread('Photos/cat.jpg')

#Large cat images. Size: 2400x1600. We need to resize it. 
#Cause it is imposible to see on it cause the image larger than my computer screen
img_2 = cv.imread('Photos/cat_large.jpg')
```

How we can show images

```python
cv.imshow('Cat', img_1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.axis("off")
plt.title("Cat")
plt.imshow(cv.cvtColor(img_1, cv.COLOR_BGR2RGB))
plt.show()
```

How we can import videos

```python
#If we use integer argument in this method, it means what we want import video through our cameras
#For ex if arg = 0, we use our WebCam
#If arg = 1/0/2 for ex, it means what we have several cameras. We can assign any argument for any camera
# capture_1 = cv.VideoCapture(0)

#Just video file from the computer
capture_2 = cv.VideoCapture('Videos/dog.mp4')
```

How we can read videos

```python
#Video file from the computer
#But with 'try' works only one time, without it gie the error when the wideo ends
while True:
    try:
        isTrue, frame = capture_2.read()
        cv.imshow('Dog', frame)

        if cv.waitKey(20) & 0xFF==ord('d'):
            cv.destroyAllWindows()
            break
            
    except: break
    
capture_2.release()
cv.destroyAllWindows()
```

```python
veb = cv.VideoCapture(0)

#Video file from the WebCam
while True:
    try:
        isTrue, frame = veb.read()
        cv.imshow('Dog', frame)

        if cv.waitKey(20) & 0xFF==ord('d'):
            cv.destroyAllWindows()
            break
            
    except: break
```
